### Importação dos dados










```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja1 = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja1.head()
```

## Características gerais da base de dados

```python
colunas_iguais = all(
    df.columns.equals(loja1.columns)
    for df in [loja2, loja3, loja4]
)

if colunas_iguais:
    print('As colunas são iguais')
else:
    print('As colunas são diferentes')
```

```python
loja1.info()
```

```python
for i, loja in enumerate([loja1, loja2, loja3, loja4], start=1):
    print(f'Loja {i}:')
    print(loja.shape)
```

```python
loja1['Loja'] = 'Loja 1'
loja2['Loja'] = 'Loja 2'
loja3['Loja'] = 'Loja 3'
loja4['Loja'] = 'Loja 4'
```

```python
dados = pd.concat([loja1, loja2, loja3, loja4])
dados.head()
```

```python
dados.info()
```

```python
dados['Frete'] = dados['Frete'].round(2)
dados['Preço'] = dados['Preço'].round(2)
dados.head()
```

#1. Análise do faturamento



```python
faturamento_total = dados['Preço'].sum()
faturamento_total
```

```python
faturamento_lojas = dados.groupby('Loja')['Preço'].sum().sort_values(ascending=False)
print(faturamento_lojas)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(9,6))
cores = ['green'] + ['gray']*(len(faturamento_lojas)-2) + ['red']

barras = plt.bar(
    faturamento_lojas.index,
    faturamento_lojas.values,
    color=cores
)

plt.title("Faturamento Total por Loja")
plt.ylabel("Faturamento")
plt.xlabel("Loja")
plt.xticks(rotation=0)

for barra in barras:
    altura = barra.get_height()
    plt.text(
        barra.get_x() + barra.get_width()/2,
        altura,
        f'{altura:,.2f}',
        ha='center',
        va='bottom'
    )

plt.show()
```

# 2. Vendas por Categoria



```python
vendas_categoria = (dados.groupby(['Loja', 'Categoria do Produto']).size().sort_values(ascending=False))
print(vendas_categoria)
```

```python
for loja in dados['Loja'].unique():
    print(f"\n{loja}")
    print(dados[dados['Loja'] == loja]['Categoria do Produto'].value_counts())
```

```python
vendas_categoria = (dados.groupby(['Loja', 'Categoria do Produto']).size().unstack())

vendas_categoria.T.plot(figsize=(10,6), marker='o')
plt.title('Distribuição de Vendas por Categoria')
plt.ylabel('Quantidade Vendida')
plt.xlabel('Categoria')
plt.xticks(rotation=45)
plt.show()
```

# 3. Média de Avaliação das Lojas


```python
avaliacao_media = (dados.groupby('Loja')['Avaliação da compra'].mean().sort_values(ascending=False))
avaliacao_media
```

```python
plt.figure(figsize=(8,5))
plt.plot(avaliacao_media.index, avaliacao_media.values, marker='o')
plt.title('Média de Avaliação por Loja')
plt.xlabel('Loja')
plt.ylabel('Avaliação Média')
plt.ylim(0,5)
plt.grid(True)
plt.show()
```

```python
comparacao = pd.DataFrame({'Faturamento': dados.groupby('Loja')['Preço'].sum(),'Avaliação Média': dados.groupby('Loja')['Avaliação da compra'].mean()})
comparacao
```

# 4. Produtos Mais e Menos Vendidos

```python
top_produtos_por_loja = (dados.groupby(['Loja', 'Produto']).size().reset_index(name='Quantidade'))
top5 = (top_produtos_por_loja.sort_values(['Loja', 'Quantidade'], ascending=[True, False]).groupby('Loja').head(5))
top5
```

```python
bottom5 = (top_produtos_por_loja.sort_values(['Loja', 'Quantidade'], ascending=[True, True]).groupby('Loja').head(5))
bottom5
```

```python
import math

produtos_vendas = (dados.groupby(['Loja', 'Produto']).size().reset_index(name='Quantidade'))

lojas = produtos_vendas['Loja'].unique()
num_lojas = len(lojas)

colunas = 2
linhas = math.ceil(num_lojas / colunas)

fig, axes = plt.subplots(linhas, colunas, figsize=(15, 5 * linhas))
axes = axes.flatten()

for i, loja in enumerate(lojas):

    dados_loja = produtos_vendas[produtos_vendas['Loja'] == loja]

    top5 = dados_loja.sort_values('Quantidade', ascending=False).head(5)
    bottom5 = dados_loja.sort_values('Quantidade', ascending=True).head(5)

    combinado = pd.concat([top5, bottom5])

    cores = ['green'] * len(top5) + ['red'] * len(bottom5)

    axes[i].barh(combinado['Produto'], combinado['Quantidade'], color=cores)
    axes[i].set_title(f'Produtos Mais e Menos Vendidos - {loja}')
    axes[i].invert_yaxis()
    axes[i].set_xlabel('Quantidade Vendida')

for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```

# 5. Frete Médio por Loja

```python
frete_medio = (dados.groupby('Loja')['Frete'].mean().reset_index().sort_values(by='Frete', ascending=False))
print(frete_medio)
```

```python
plt.figure(figsize=(9,6))

cores = ['red'] + ['gray']*(len(frete_medio)-2) + ['green']

barras = plt.bar(frete_medio['Loja'], frete_medio['Frete'], color=cores)
plt.title('Frete Médio por Loja')
plt.ylabel('Frete Médio')
plt.xlabel('Loja')
plt.xticks(rotation=45)

for barra in barras:
    altura = barra.get_height()
    plt.text(
        barra.get_x() + barra.get_width()/2,
        altura,
        f'{altura:.2f}',
        ha='center',
        va='bottom'
    )

plt.show()
```

```python
frete_medio = dados.groupby('Loja')['Frete'].mean()

avaliacao_media = dados.groupby('Loja')['Avaliação da compra'].mean()

comparacao = pd.concat([frete_medio, avaliacao_media], axis=1).reset_index()

plt.figure(figsize=(7,4))
plt.scatter(comparacao['Frete'], comparacao['Avaliação da compra'])

for i in range(len(comparacao)):
    plt.text(comparacao['Frete'][i], comparacao['Avaliação da compra'][i], comparacao['Loja'][i])

plt.title('Frete Médio vs Avaliação Média')
plt.xlabel('Frete Médio')
plt.ylabel('Avaliação Média')
plt.show()
```

# 📊 RELATÓRIO FINAL — ANÁLISE ESTRATÉGICA DAS LOJAS DO SENHOR JOÃO

## 1️⃣ Introdução

Este relatório tem como objetivo avaliar o desempenho das quatro unidades comerciais do Senhor João e indicar, com base em dados concretos, qual delas apresenta menor viabilidade estratégica para permanência no portfólio.

Foram analisados os seguintes indicadores:

* 📈 Faturamento total
* 📦 Categorias mais e menos vendidas
* 🛒 Produtos com maior e menor saída
* ⭐ Média de avaliação dos clientes
* 🚚 Frete médio por loja

A decisão final considera desempenho financeiro, competitividade comercial e potencial de crescimento.

---

# 2️⃣ Desenvolvimento da Análise

## 📊 2.1 Faturamento Total

A análise do faturamento revelou que, embora os valores sejam relativamente próximos, existe um ranking claro de desempenho:

* 🥇 Loja 1 — Maior faturamento
* 🥈 Loja 2
* 🥉 Loja 3
* ❌ Loja 4 — Menor faturamento

Mesmo com diferenças não muito expressivas, a Loja 4 permanece consistentemente atrás das demais. Isso indica menor geração de receita e menor impacto financeiro dentro do conjunto analisado.

**Conclusão parcial:**
Em termos de lucratividade, a Loja 4 é a menos eficiente.

---

## 📦 2.2 Categorias de Produtos

A análise por categoria mostrou que:

* Lojas com maior diversidade e força em categorias estratégicas tendem a gerar maior faturamento.
* A Loja 1 apresenta forte desempenho em categorias de alto valor agregado.
* A Loja 4 é a única que ficou abaixo da média geral em desempenho por categoria.

Isso demonstra menor competitividade estrutural da Loja 4.

**Conclusão parcial:**
Menor força comercial por categoria impacta diretamente o resultado final.

---

## 🛒 2.3 Produtos Mais e Menos Vendidos

A análise de volume de produtos revelou:

* A Loja 2 lidera em produtos campeões de venda.
* A Loja 3 possui maior concentração de produtos com baixa saída.
* A Loja 4 apresenta desempenho intermediário, mas sem produtos líderes expressivos.

Além disso, lojas com muitos produtos de baixa saída tendem a sofrer com imobilização de estoque e menor giro de capital.

**Conclusão parcial:**
A Loja 4 não demonstra força comercial significativa em produtos estratégicos.

---

## ⭐ 2.4 Avaliação Média dos Clientes

A satisfação do cliente é um indicador fundamental de sustentabilidade no longo prazo.

Médias registradas:

* 🥇 Loja 3 — 4,05
* 🥈 Loja 2 — 4,04
* 🥉 Loja 4 — 4,00
* ❌ Loja 1 — 3,98

Embora a Loja 4 não tenha a pior avaliação, também não se destaca positivamente. Ela ocupa uma posição mediana.

**Conclusão parcial:**
Não há diferencial competitivo relevante em satisfação do cliente.

---

## 🚚 2.5 Frete Médio

Valores médios:

* Loja 1 — R$ 34,69
* Loja 2 — R$ 33,62
* Loja 3 — R$ 33,07
* Loja 4 — R$ 31,28

A Loja 4 possui o menor frete médio.

Entretanto, mesmo com esse benefício competitivo, não conseguiu converter essa vantagem em maior faturamento.

**Conclusão parcial:**
O frete mais baixo não foi suficiente para impulsionar o desempenho financeiro.

---

# 3️⃣ Análise Estratégica Comparativa

Ao observar o desempenho geral:

* 🔹 Loja 1 lidera em faturamento.
* 🔹 Loja 2 apresenta equilíbrio entre receita e avaliação.
* 🔹 Loja 3 possui a melhor avaliação média.
* 🔹 Loja 4 apresenta menor faturamento e menor destaque competitivo.

A Loja 4 não lidera em nenhum indicador relevante.

Ela não é a pior em todos os aspectos — mas também não é a melhor em nenhum.

Isso caracteriza ausência de diferencial estratégico.

---

# 4️⃣ Conclusão e Recomendação

Considerando todos os fatores analisados — financeiros, comerciais e operacionais — recomenda-se que o Senhor João venda a Loja 4.

### Justificativas principais:

✔ É a loja com menor faturamento total.
✔ Não apresenta liderança em categorias estratégicas.
✔ Não possui produtos campeões de venda expressivos.
✔ Possui avaliação apenas mediana.
✔ O frete mais baixo não compensou o desempenho inferior.

---

# 📌 Recomendação Final

Portanto, para otimização do portfólio e maximização de resultados, a recomendação técnica é sua alienação (venda), mantendo-se as Lojas 1, 2 e 3, que apresentam maior solidez e potencial de crescimento.
